```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#pulls 1000 random numbers
n = 1000
x = np.random.rayleigh(scale=10,size=n)

#puts the numbers into 100 bins and changes the appearance of them
plt.hist(x, bins=100,alpha= 0.7,edgecolor='black')

#labeling axes again
plt.xlabel("x")
plt.ylabel("n per bin")

#saving it!!!
plt.savefig("rayleigh_graph.pdf", bbox_inches='tight', facecolor='white')

plt.show()
```
