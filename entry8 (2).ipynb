```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#first we make an array
x = np.linspace(0,1,100)

#now we tell it to define sin(x) and cos(x)
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
#multipanel plot coming right up!
f, axarr = plt.subplots(1,2)

#here, we plot the sin graph and label axes
axarr[0].plot(x,sin(x))
axarr[0].set_title('sin(x)')
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')

#here, we do the same for cos
axarr[1].plot(x,cos(x))
axarr[1].set_title('cos(x)')
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')

#saving and making sure it looks nice
plt.savefig('entry8.pdf',bbox_inches='tight', facecolor='white')

#need to add more space so we can see the y-axis cos(x)
f.subplots_adjust(wspace=0.5)

plt.show()
```
