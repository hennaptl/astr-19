```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exp_x(x):
    return np.exp(x)
    
#this creates the array we want
x = np.linspace(0,1,100)
```

```python
y = exp_x(x)

#making the actual figure
fig=plt.figure(figsize=(7,7))
plt.plot(x,y)

#now to define the axes
plt.xlabel="Time[milliseconds]"
plt.ylabel="Awesomeness"

#now we save as PDF under the following conditions!
plt.savefig("session7.pdf", bbox_inches="tight", facecolor='white')
plt.show()
```
